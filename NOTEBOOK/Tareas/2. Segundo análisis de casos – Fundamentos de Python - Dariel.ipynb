# Problema #1: Calculadora de propina inteligente

## Problema

Solicitar al usuario que dijite el monto total de una cuenta y el porcentaje que desea dejar, el programa debe de calcular el valor de la propina a dejar y el total final pagar. Se deben validar los datos ingresados sean numéricos utilizando la estuctura try - except


## Entrada
- Monto a pagar
El porcentaje de propina a dejar

## Proceso
- Solicitar al usuario lo siguiente:
Monto de la cuenta
Porcentaje de propina deseado a dejar 
Que los datos ingresados sean float
Calcular la propina con la fórmula
Propina = Monto * Porcentaje / 100
Calcular el total a pagar = Monto + Propina
Mostrar los resultados
Si el valor no es numérico mostrar "Error, no es un valor numérico"

## Salida esperada
-Valor de la propina 
Total a pagar 

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
|Caso normal| ₡ 15500 y 10     |   ₡17050 y ₡1550   |  ₡17050 y ₡1550    |
|Caso límite| ₡ 0  y 10        |  ₡0 y ₡0           |   ₡0 y ₡0          |
|Caso inválido|quince mil quinientos|   Error       |   Error            |


```python
try:
    monto = float(input("Introduzca el monto de la cuenta: "))
    porcentaje = float(input("Introduzca el porcentaje que desea dar de propina: "))
    propina = monto * (porcentaje / 100) 
    total_a_pagar = monto + propina
    print(f"Valor de la propina: ₡{propina: .2f}")
    print(f"Total a pagar: ₡{total_a_pagar: .2f}")
except ValueError: 
    print("Error, no es un valor numérico")
```

# Problema #2: Detector de número par/impar

## Problema
Identificar los números como par o impar

## Entrada
-Ingresar un número entero

## Proceso
- Pedir al usuario que ingrese un número entero
Que los datos inresados sean int
Utilizar % para tener el residuo de la división entre 2
residuo = 0, el número es par
residuo ≠ 0, el número es impar
Si se ingresa un valor que no sea un número, mostrar mensaje de error

## Salida esperada
El número es par/impar
En caso de error, "Error, no es un valor numérico"

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal|  8              |    Es par          |     Es par         |
| Caso límite|  -5             |    Es impar        |     Es impar       |
| Caso inválido|  Tres         |    Error           |     Error          |


```python
try: 
    numero = int(input("Ingrese un número entero: "))
    if numero % 2 == 0:
        print(f"El número {numero} es par") 
    else:
        print(f"El número {numero} es impar")
except ValueError:
    print("Error, no es un valor numérico")        
```

# Problema #3: Clasificador de edad

## Problema
Solicitar la edad al usuario y calificarla en menor, adulto y adulto mayor.

## Entrada
-Ingrese la edad 
## Proceso 
- Se solicita que ingrese a edad 
Que el dato sea un int Si la edad ingresada < 18, es menor de edad, si la edad ingresada es ≥ 18 y < 65, es adulto 
Si la edad ingresada es ≥ 65, es adulto mayor Mostrar resultados 
Si se ingresa un valor no numérico o no entero, mostrar mensaje de error 
## Salida esperada 
- Es: 
Menor de edad 
Adulto 
Adulto mayor 
Error "No es una edad válida"


## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal|  15             |  Menor de edad     | Menor de edad      |
| Caso límite|  18             |   Adulto           |   Adulto           |
| Caso inválido|  -5           |   No es válida     | No es válida       |


```python
try:
    edad = int(input("Ingrese la edad: "))
    if edad < 0:
        print("No es una edad válida")
    elif edad < 18:
        print("Es menor de edad")
    elif edad >= 18 and edad < 65: 
        print("Es adulto")   
    elif edad >= 65:
        print("Es adulto mayor")
except ValueError:
    print("No es una edad válida")
```

# Problema #4: Validador de compra con descuento

## Problema

Pedir al usuario el monto de la compra y si es cliente frecuente, se debe determinar si se le aplica descuneto y mostrar el monto a pagar

## Entrada
- Ingrese el nombre del cliente
Monto de la compra
Es cliente frecuente? si/no

## Proceso
- Solicitar el nombre del cliente
Solicitar el monto
Que el monto sea float
Solicitar si el cliente es frecuente
Si el monto es mayor a 30000 y el cliente es frecuente, el descuento es de 25%
Si el monto es mayor a 30000 y el cliente no es frecuente, el descuento es de 5%
Si el monto es menor a 30000 y el cliente es frecuente, el descuento es de 10%
Si el monto es menor a 30000 y el cliente no es frecuente, no hay descuento
Calcular el descuento
Calcular el total a pagar
Mostrar resutados y si se ingresa un valor no válido, mostrar un mensaje de error

## Salida esperada
- Nombre del cliente
Descuento aplicado
Total a pagar
En caso de error "No es un monto válido"

## Pruebas manuales

| Escenario | Datos de entrada    | Resultado esperado | Resultado obtenido |
|-----------|------------------   |--------------------|--------------------|
| Caso normal |  Dariel, 30000, si| Dariel, 3000, 27000| Dariel, 3000, 27000|
| Caso límite | Dariel, 30000, no | Dariel, 0, 30000   | Dariel, 0, 30000   |
| Caso inválido| Dariel, no, no   | Resultado inválido |  Resultado inválido|


```python
try:
    cliente = input("Ingrese el nombre del cliente: ")
    monto = float(input("Ingrese el monto: "))
    cliente_frecuente = input("¿Es cliente frecuente?: ")
    if monto > 30000 and cliente_frecuente == "si":
        descuento = monto * 0.25
    elif monto > 30000 and cliente_frecuente == "no": 
        descuento = monto * 0.05
    elif monto <= 30000 and cliente_frecuente == "si":
        descuento = monto * 0.10
    else:
        descuento = 0 
    total_a_pagar = monto - descuento  
    print(f"Cliente: {cliente}")  
    print(f"Descuento aplicado: ₡{descuento:.2f}")  
    print(f"Total a pagar: ₡{total_a_pagar:.2f}")  
except ValueError:
   print("No es un monto válido") 
        
        
```
